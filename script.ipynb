```python
import os
```

```python
#Setting some variables
file_name = "put a txt file name here"
Desktop_Path = "put the path to your desktop"
initial_dir_path = os.getcwd()

#obtaining the ids of the dangling blobs(files deleted by git)
os.chdir(Desktop_Path)
os.system("git fsck > "+initial_dir_path+"/"+file_name)
```

```python
os.chdir(initial_dir_path)

# oppening and reading path
my_file = open(file_name, "r")
data = my_file.read()

#isolating the ids and putting them in a list
blob_list = data.split("\n")
for i in range(len(blob_list)):
    blob_list[i] = blob_list[i][14:]

print(blob_list[0])
```

```python
# Making an Empty Directory to store files
os.mkdir(initial_dir_path+"/recoveredfiles")
```

```python
#recovering the files
counter = 1
feedstring = ""
for i in blob_list:
    feedstring = "git show "+str(i)+" > "+initial_dir_path+"/recoveredfiles/file"+str(counter)
    print(feedstring)
    os.system(feedstring)
    counter += 1
```
